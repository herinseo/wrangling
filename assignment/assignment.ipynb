```python
import pandas as pd
import numpy as numpy
import seaborn as sns
```

```python
# 2.1
df = pd.read_csv('./data/airbnb_hw.csv', low_memory=False)
print(df.shape, '\n')
df.head()
```

```python
# since imported as a string, must elimintae the commas

price = df['Price']
price = price.str.replace(',', '')
# typecast price from string to float/numeric
price = pd.to_numeric(price, errors='coerce')
print(price.unique(), '\n')
print('Total missing values', sum(price.isnull()))
```

```python
# 2.2
df_sharks = pd.read_csv('./data/sharks.csv', low_memory = False)
df_sharks.head()
```

```python
df_sharks['Type'].value_counts()
```

```python
import numpy as np
type = df_sharks['Type']
# grouping similar types into one type variable -- eliminates redundancy
type = type.replace(['Sea Disaster','Boat','Boating','Boatomg'],'Watercraft')
type = type.replace(['Invalid','Questionable','Unconfirmed','Unverified','Under investigation'],np.nan)

type.value_counts()
df_sharks['Type'] = type
del type
df_sharks['Type'].value_counts()
```

```python
# 2.3

# use of chatgpt here to read in .parquet file
import pandas as pd

df_justice = pd.read_parquet('./data/justice_data.parquet')
print(df_justice.head())
```

```python
import numpy as np
release = df_justice['WhetherDefendantWasReleasedPretrial']
# value 9 means unclear, so eliminate by changing to na
release = release.replace(9, np.nan)
print(release.value_counts, '\n')
sum(release.isnull())
df_justice['WhetherDefendantWasReleasedPretrial'] = release
del release
```

```python
# 2.4
sentence = df_justice['ImposedSentenceAllChargeInContactEvent'] # some values are just ' ' need to change to na value
type = df_justice['SentenceTypeAllChargesAtConvictionInContactEvent']

sentence = pd.to_numeric(sentence, errors='coerce')
sentence_na = sentence.isnull()
print(np.sum(sentence_na),'\n') # 9k missing out of 23k, a lot of missing data. need to clean to have much less missing

print(pd.crosstab(sentence_na, type),'\n')
sentence = sentence.mask(type==4, 0)
sentence = sentence.mask(type==9, np.nan)

sentence_na = sentence.isnull()
print(pd.crosstab(sentence_na, type), '\n')
print(np.sum(sentence_na),'\n')

df_justice['ImposedSentenceAllChargeInContactEvent'] = sentence
del sentence, type

# now have 274 missing which is much more reasonable
```
